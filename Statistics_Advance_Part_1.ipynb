"Regardless of the shape of the population distribution, the distribution of the sample mean (or sum) approaches a normal distribution as the sample size increases, provided the samples are independent and identically distributed (i.i.d.) and the population has a finite variance."

Justifies the Use of the Normal Distribution:

Foundation for Inferential Statistics:

Simplifies Analysis:

 Many statistical methods (like t-tests, ANOVA, and regression) rely on the assumption of normality, which CLT helps satisfy in large samples.


Works with Different Distributions:

The population can be skewed, uniform, or even multimodal—CLT ensures that the sampling distribution of the mean will still be approximately normal.

```python
#Answer 3

def bernoulli_pdf(x, p):
    """
    Computes the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The value (0 or 1) for which to compute the probability.
    p (float): The probability of success (0 ≤ p ≤ 1).

    Returns:
    float: The probability P(X = x).
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage:
p = 0.7  # Probability of success
print("P(X=0):", bernoulli_pdf(0, p))
print("P(X=1):", bernoulli_pdf(1, p))
```

```python
#Answer 9

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    """
    Computes Z-scores for a given dataset.

    Parameters:
    data (array-like): Input dataset.

    Returns:
    np.ndarray: Z-scores of the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Use sample standard deviation
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(z_scores):
    """
    Plots the Z-scores as a histogram.

    Parameters:
    z_scores (array-like): Computed Z-scores.
    """
    plt.figure(figsize=(8, 5))
    sns.histplot(z_scores, bins=30, kde=True, color='blue', edgecolor='black')

    # Plot standard normal curve for reference
    x = np.linspace(-4, 4, 100)
    y = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x**2)
    plt.plot(x, y, 'r-', label="Standard Normal Curve")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-score Distribution")
    plt.axvline(0, color='black', linestyle='dashed', label="Mean (Z=0)")
    plt.legend()
    plt.grid(alpha=0.5)

    plt.show()

# Example usage
np.random.seed(42)
data = np.random.normal(50, 10, 1000)  # Generate normal data (mean=50, std=10)
z_scores = calculate_z_scores(data)
plot_z_scores(z_scores)
```

```python
#Answer 19

import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a one-sample Z-test for hypothesis testing.

    Parameters:
    sample (array-like): Sample data.
    pop_mean (float): Population mean (H0).
    pop_std (float): Population standard deviation.
    alpha (float): Significance level (default=0.05).

    Returns:
    tuple: (Z-score, p-value, decision)
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Decision: Reject or Fail to Reject H0
    decision = "Reject H0" if p_value < alpha else "Fail to Reject H0"

    return z_score, p_value, decision

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(52, 10, 50)  # Sample (n=50) from normal distribution (mean=52, std=10)
pop_mean = 50  # Hypothesized population mean
pop_std = 10   # Population standard deviation

# Perform Z-test
z_score, p_value, decision = z_test(sample_data, pop_mean, pop_std)

# Print Results
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.5f}")
print(f"Decision: {decision}")
```

```python
#Answer 20

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a dataset.

    Parameters:
    data (array-like): Sample data.
    confidence (float): Confidence level (default=0.95 for 95%).

    Returns:
    tuple: (Lower bound, Upper bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error: (sample std / sqrt(n))

    # Compute critical Z-value for the given confidence level
    z_critical = stats.norm.ppf((1 + confidence) / 2)

    # Compute the confidence interval
    margin_of_error = z_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(50, 10, 100)  # Generate 100 values (mean=50, std=10)

# Compute 95% Confidence Interval
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

# Print Results
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Answer 21

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a dataset.

    Parameters:
    data (array-like): Sample data.
    confidence (float): Confidence level (default=0.95 for 95%).

    Returns:
    tuple: (Lower bound, Upper bound, Sample Mean)
    """
    n = len(data)
    sample_mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error (sample std / sqrt(n))

    # Compute critical Z-value for the confidence level
    z_critical = stats.norm.ppf((1 + confidence) / 2)

    # Compute confidence interval
    margin_of_error = z_critical * std_err
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean

# Step 1: Generate Normally Distributed Data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, n=100

# Step 2: Compute 95% Confidence Interval
ci_lower, ci_upper, sample_mean = confidence_interval(sample_data, confidence=0.95)

# Step 3: Visualization
plt.figure(figsize=(8, 5))
plt.hist(sample_data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(ci_lower, color='red', linestyle='dashed', label=f"Lower CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', label=f"Upper CI = {ci_upper:.2f}")
plt.axvline(sample_mean, color='blue', linestyle='solid', label=f"Sample Mean = {sample_mean:.2f}")

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Data with 95% Confidence Interval")
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# Step 4: Print Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Answer 23

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define Poisson distribution parameter (lambda)
lambda_value = 4  # Expected occurrences (e.g., avg = 4 events per interval)

# Step 2: Generate x values (possible event counts)
x = np.arange(0, 15)  # Consider event counts from 0 to 15

# Step 3: Compute CDF values
cdf_values = poisson.cdf(x, lambda_value)

# Step 4: Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='mid', color='blue', label=f"Poisson CDF (λ={lambda_value})", linewidth=2)
plt.xlabel("Number of Events (k)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of a Poisson Distribution")
plt.axhline(0.95, color='red', linestyle='dashed', label="95% Probability Threshold")
plt.legend()
plt.grid(alpha=0.5)

# Show the plot
plt.show()

# Step 5: Interpretation - Probability of k ≤ 5
k_value = 5
probability = poisson.cdf(k_value, lambda_value)
print(f"P(X ≤ {k_value}) = {probability:.4f} (Probability of 5 or fewer events)")
```

```python
#Answer 26

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a Normally Distributed Dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Size=1000

# Step 2: Compute Range & Interquartile Range (IQR)
data_range = np.max(data) - np.min(data)
q1 = np.percentile(data, 25)  # First quartile (Q1)
q3 = np.percentile(data, 75)  # Third quartile (Q3)
iqr = q3 - q1  # IQR Calculation

# Step 3: Visualization
plt.figure(figsize=(8, 5))
sns.boxplot(x=data, color="lightblue")

# Mark Q1, Q3, and IQR range
plt.axvline(q1, color='red', linestyle='dashed', label=f"Q1 (25th percentile) = {q1:.2f}")
plt.axvline(q3, color='green', linestyle='dashed', label=f"Q3 (75th percentile) = {q3:.2f}")
plt.axvline(np.median(data), color='blue', linestyle='solid', label=f"Median = {np.median(data):.2f}")

plt.xlabel("Value")
plt.title("Boxplot Showing IQR of Normal Distribution")
plt.legend()
plt.grid(alpha=0.5)

# Show the plot
plt.show()

# Step 4: Print Results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#Answer 27

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a Normally Distributed Dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std=15

# Step 2: Compute Z-score Normalization
mean = np.mean(data)
std_dev = np.std(data)
normalized_data = (data - mean) / std_dev  # Apply Z-score formula

# Step 3: Plot Before & After Normalization
plt.figure(figsize=(10, 5))

# Original Data Distribution
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', label=f"Mean = {mean:.2f}")
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Normalized Data Distribution
plt.subplot(1, 2, 2)
sns.histplot(normalized_data, bins=30, kde=True, color='green', alpha=0.7)
plt.axvline(0, color='red', linestyle='dashed', label="Mean = 0")
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Step 4: Print Summary Statistics
print(f"Original Data -> Mean: {mean:.2f}, Std Dev: {std_dev:.2f}")
print(f"Normalized Data -> Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data):.2f}")
```

```python
#Answer 28

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

def analyze_distribution(size=1000, mean=50, std_dev=10):
    # Step 1: Generate Normally Distributed Data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Compute Skewness & Kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # Excess kurtosis (subtracts 3)

    # Step 3: Visualization
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color="skyblue", edgecolor="black")
    plt.axvline(np.mean(data), color='red', linestyle='dashed', label=f"Mean = {np.mean(data):.2f}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Histogram of Normal Distribution")
    plt.legend()
    plt.grid(alpha=0.5)

    # Show the plot
    plt.show()

    # Step 4: Print Results
    print(f"Skewness: {skewness:.4f} (Expected ≈ 0 for Normal Distribution)")
    print(f"Kurtosis: {kurt:.4f} (Excess Kurtosis, Expected ≈ 0 for Normal Distribution)")

# Run the function
analyze_distribution()
```
